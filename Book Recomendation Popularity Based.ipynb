## Model Building

```python
print(books.columns)
print(raitings.columns)
```

```python
data=raitings.merge(books,on="ISBN")
data.head()
```

```python
data.shape
```

```python
num_data=data.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_data.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_data
```

```python
avg_rating = data.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating': 'Avg_rating'}, inplace=True)
avg_rating
```

```python
popular=num_data.merge(avg_rating,on='Book-Title')
popular
```

```python
popular=popular[popular['num_rating']>=250].sort_values('Avg_rating',ascending=False).head(50)
```
