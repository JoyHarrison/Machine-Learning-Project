## Model Building

```python
rating=ratings.merge(books,on='ISBN')
rating
```

```python
X=rating.groupby('User-ID').count()['Book-Rating']> 200
well_read=X[X].index
well_read
```

```python
fill_rating=rating[rating['User-ID'].isin(well_read)]
fill_rating
```

```python
y=fill_rating.groupby('Book-Title').count()['Book-Rating']>=50
pop_book=y[y].index
pop_book
```

```python
final_rating=fill_rating[fill_rating['Book-Title'].isin(pop_book)]
final_rating
```

```python
final_rating.drop_duplicates()
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```
