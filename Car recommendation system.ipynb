```python
data['fuel_type'].fillna("Petrol", inplace=True)
```

```python
data['fuel_type'].isnull().sum()
```

```python
data['kms_driven'].dropna(inplace=True)
data['kms_driven'].isnull().sum()
```

```python
data.dropna(subset=['kms_driven'], inplace=True)
data['kms_driven'].isnull().sum()
```

```python
data['Price'] = data['Price'].replace({'Ask For Price': None, ',': '', ' ': ''}, regex=True)
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
```

```python
data['kms_driven'] = data['kms_driven'].str.replace(' kms', '').str.replace(',', '')
data['kms_driven'] = pd.to_numeric(data['kms_driven'], errors='coerce')
```

```python
data
```

```python
data['year'] = pd.to_numeric(data['year'], errors='coerce')
data
```

```python
# Recommendation Function
def recommend_cars(company, year, price, kms_driven, fuel_type, tolerance=0.2):
    # Filter the data based on company and fuel type
    filtered_data = data[(data['company'].str.lower() == company.lower()) & 
                         (data['fuel_type'].str.lower() == fuel_type.lower())]
    
    # Further filter based on year, price, and kms_driven with some tolerance
    recommended_cars = filtered_data[
        (filtered_data['year'] >= year * (1 - tolerance)) & 
        (filtered_data['year'] <= year * (1 + tolerance)) &
        (filtered_data['Price'] >= price * (1 - tolerance)) & 
        (filtered_data['Price'] <= price * (1 + tolerance)) &
        (filtered_data['kms_driven'] >= kms_driven * (1 - tolerance)) & 
        (filtered_data['kms_driven'] <= kms_driven * (1 + tolerance))
    ]
    
    return recommended_cars[['name', 'company', 'year', 'Price', 'kms_driven', 'fuel_type']]

recommend_cars(company='Hyundai', year=2014, price=325000, kms_driven=30000, fuel_type='Petrol')

```
